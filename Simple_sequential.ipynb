```
import numpy as np
import keras
import tensorflow as tf
import pandas as pd
import keras.backend as K
K.clear_session()
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout
```

```
url = 'https://github.com/fiedler-m/pitch_predictor/raw/main/scherzer_HardSoft.csv'
data = pd.read_csv(url)

df = pd.DataFrame(data)
df
```

```
pitches = df['pitch_type']
pitches = pd.get_dummies(pitches) #hot encoding for pitches
p_enc = pitches.to_numpy()

del df['pitch_type']

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

#right or left handed
df["stand"] = le.fit_transform(df["stand"])

ty = pd.get_dummies(df['prev_type'])
df = df.drop('prev_type',axis = 1)
df = df.join(ty)

#hot encode previous pitch string
prevs = pd.get_dummies(df['prev_pitch'])
df = df.drop('prev_pitch',axis = 1)

#Join encoded df
df = df.join(prevs)
variables = df.to_numpy()

variables
```

```
from sklearn.model_selection import train_test_split
trainX, testX, trainY, testY = train_test_split( 
    variables, p_enc, test_size=0.2)
```

```
model = Sequential()
from keras.optimizers import SGD
model.add(Dense(8, activation = 'relu', input_dim = 13))
model.add(Dense(3, activation = 'softmax'))

model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

model.fit(trainX, trainY, epochs = 100,
            batch_size = 64)
```

```
model.fit(trainX, trainY, epochs = 200,
            batch_size = 64)

score = model.evaluate(testX, testY, batch_size = 64)
print("Test Accuracy: " + "{:.2%}".format(score[1]))
```
